### Group collisions by date and build overview trend chart

```python
collisions_by_date = df_collisions_all.groupby(['COLLISION_DATE_T'])['CASE_ID'].aggregate(['count'])
collisions_by_date = collisions_by_date.reset_index()
```

```python
line_chart = alt.Chart(collisions_by_date).mark_line(tooltip=True, size=1).encode(
    x=alt.X(
        'COLLISION_DATE_T:T',
        title="Date"
    ),
    y=alt.Y(
        'sum(count):Q',
         title="Number of Collisions"
        
    )
).properties(
    width=800,
    height=300,
    title = "Trend of California Traffic Collisions Over the Past 10 Years"
).interactive()

point_chart = alt.Chart(collisions_by_date).mark_point(tooltip=True, size=50, opacity=0).encode(
    x=alt.X(
        'COLLISION_DATE_T:T',
        title="Date"
    ),
    y=alt.Y(
        'sum(count):Q',
         title="Number of Collisions"
        
    )
).properties(
    width=800,
    height=300,
).interactive()

line_chart = line_chart+point_chart.add_selection(alt.selection_single())
line_chart
```

### Trend and % of Collisions chart

#### Method 1. Use Altair binding_checkbox for displaying choices

```python
# create a checkbox selection for alcohol involved cases
check_alcohol = alt.binding_checkbox(name="% of Collisions Involving Alcohol")
alcohol_selection = alt.selection_single(bind=check_alcohol, name="Alcohol")
alcohol_checkbox_condition = alt.condition(alcohol_selection,
                                        alt.value(0.0),
                                        alt.value(1.0)
                                       )

# prepare data for alcohol involved collisions
df_alcohol = df_collisions_all.loc[df_collisions_all['ALCOHOL_INVOLVED'] == 'Y']
alcohol_by_date = df_alcohol.groupby(['COLLISION_DATE_T'])['CASE_ID'].aggregate(['count'])
alcohol_by_date = alcohol_by_date.reset_index()

df_collisions_alcohol = collisions_by_date
df_collisions_alcohol = df_collisions_alcohol.rename(columns={"COLLISION_DATE_T": "Date", "count": "Total Collisions"})
df_collisions_alcohol['Collisions Involving Alcohol'] = alcohol_by_date['count']
df_collisions_alcohol['Alcohol'] = df_collisions_alcohol['Collisions Involving Alcohol'] / df_collisions_alcohol["Total Collisions"]

alcohol_line = alt.Chart(df_collisions_alcohol).mark_line(tooltip=True, color="red").encode(
    x=alt.X(
        'year(Date):O',
    ),
    y=alt.Y(
        'mean(Alcohol):Q',
        axis=alt.Axis(format='%'),
        title="% of Collisions"
    ),
    opacity=alcohol_checkbox_condition
).properties(
    width=800,
    height=300,
    title = ""
)
alcohol_points = alt.Chart(df_collisions_alcohol).mark_point(tooltip=True, color="red").encode(
    x=alt.X(
        'year(Date):O',
    ),
    y=alt.Y(
        'mean(Alcohol):Q',
        axis=alt.Axis(format='%'),
        title="% of Collisions"
    ),
    opacity=alcohol_checkbox_condition
)

alcohol_chart = alcohol_line+alcohol_points
alcohol_chart = alcohol_chart.add_selection(alcohol_selection)
```

```python
# prepare data for Motor Cycle involved collisions
df_motorcycle = df_collisions_all.loc[df_collisions_all['MOTORCYCLE_ACCIDENT'] == 'Y']
motorcycle_by_date = df_motorcycle.groupby(['COLLISION_DATE_T'])['CASE_ID'].aggregate(['count'])
motorcycle_by_date = motorcycle_by_date.reset_index()

df_collisions_motorcycle = collisions_by_date
df_collisions_motorcycle = df_collisions_motorcycle.rename(columns={"COLLISION_DATE_T": "Date", "count": "Total Collisions"})
df_collisions_motorcycle['Collisions Involving MotorCycle'] = motorcycle_by_date['count']
df_collisions_motorcycle['Motor Cycles'] = df_collisions_motorcycle['Collisions Involving MotorCycle'] / df_collisions_motorcycle["Total Collisions"]
```

```python
# prepare data for Truck involved collisions
df_truck = df_collisions_all.loc[df_collisions_all['TRUCK_ACCIDENT'] == 'Y']
truck_by_date = df_truck.groupby(['COLLISION_DATE_T'])['CASE_ID'].aggregate(['count'])
truck_by_date = truck_by_date.reset_index()

df_collisions_truck = collisions_by_date
df_collisions_truck = df_collisions_truck.rename(columns={"COLLISION_DATE_T": "Date", "count": "Total Collisions"})
df_collisions_truck['Collisions Involving Trucks'] = truck_by_date['count']
df_collisions_truck['Trucks'] = df_collisions_truck['Collisions Involving Trucks'] / df_collisions_truck["Total Collisions"]
```

```python
df_collisions_by_type = df_collisions_alcohol
df_collisions_by_type['Pedestrians'] = df_collisions_pedestrian['Pedestrians']
df_collisions_by_type['Bicycles'] = df_collisions_bicycle['Bicycles']
df_collisions_by_type['Motor Cycles'] = df_collisions_motorcycle['Motor Cycles']
df_collisions_by_type['Trucks'] = df_collisions_truck['Trucks']

df_collisions_by_type
```

```python
df_collisions_by_type1 = df_collisions_by_type.reset_index().melt(id_vars=['Date'], 
                                                                  value_vars=['Alcohol', 'Pedestrians', 'Bicycles', 'Motor Cycles', 'Trucks'], 
                                                                  var_name='% of Collisions Involving',
                                                                  value_name='pct')

df_collisions_by_type1['pct'] = pd.Series([round(val, 4) for val in df_collisions_by_type1['pct']])
df_collisions_by_type1
```

```python
# define selection
df_collision_type = pd.DataFrame({'% of Collisions Involving': ['Alcohol', 'Pedestrians', 'Bicycles', 'Motor Cycles', 'Trucks']})

selection = alt.selection_multi(fields=['% of Collisions Involving'], empty='all')
color = alt.condition(selection, alt.Color('% of Collisions Involving:N', 
                                           legend=None, 
                                           scale=alt.Scale(scheme="category10")
                                          ), 
                      alt.value('lightgray'))
type_selector = alt.Chart(df_collision_type).mark_circle(tooltip=True, size=400).encode(
    y=alt.Y('% of Collisions Involving', title=""), 
    color=color,
    tooltip=[alt.Tooltip('% of Collisions Involving', title="Collisions Involving")
            ]
).add_selection(selection).properties(title="Select to View % of Collisions")

# percentage collisions charts
pct_line_chart = alt.Chart(df_collisions_by_type1).mark_line(point=True).encode(
    x='year(Date):O',
    y=alt.Y('mean(pct):Q', title="% of Collisions", axis=alt.Axis(format='%')),
    color=color
).transform_filter(selection).add_selection(selection)

pct_point_chart = alt.Chart(df_collisions_by_type1).mark_point(tooltip=True, size=50, opacity=0).encode(
    x='year(Date):O',
    y=alt.Y('mean(pct):Q', title="% of Collisions", axis=alt.Axis(format='%')),
    color=color,
    tooltip = [
        alt.Tooltip('year(Date):O', title='Year'),
        alt.Tooltip('mean(pct):Q', title="% of Collisions", format='%')
        ]
).transform_filter(selection).add_selection(alt.selection_single())
    
# total collisions charts
collisions_line_chart = alt.Chart(collisions_by_date).mark_line(point=True, color='black').encode(
    x=alt.X(
        'year(COLLISION_DATE_T):O',
        title="Year",
    ),
    y=alt.Y(
        'sum(count):Q',
         title="Number of Collisions"
    )
).add_selection(alt.selection_single())

collisions_point_chart = alt.Chart(collisions_by_date).mark_point(size=50, opacity=0).encode(
    x=alt.X(
        'year(COLLISION_DATE_T):O',
        title="Year"
    ),
    y=alt.Y(
        'sum(count):Q',
         title="Number of Collisions"
    ),
    tooltip = [
        alt.Tooltip('year(COLLISION_DATE_T):O', title='Year'),
        alt.Tooltip('sum(count):Q', title="Number of Collisions")
    ],
).add_selection(alt.selection_single())

total_collisions_chart = collisions_line_chart + collisions_point_chart
pct_charts = pct_line_chart + pct_point_chart

type_selector | pct_charts
```

```python
layered_chart1 = alt.layer(
    total_collisions_chart,
    pct_charts
).properties(
    width=600,
    height=300,
    title="Trend and % of Collisions for Various Types"
).resolve_scale(
    y = 'independent'
).add_selection(alt.selection_single())

percent_collisions = alt.hconcat(layered_chart1, type_selector)
percent_collisions = percent_collisions.configure_axis(grid=True)

display(percent_collisions)

display(HTML("""
<p style="text-align:right">Use <b>Shift+Click</b> for Multi Selection</p>
"""))
```
